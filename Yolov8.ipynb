```python
import torch

def get_gpu_name():
    if torch.cuda.is_available():
        gpu_name = torch.cuda.get_device_name(torch.cuda.current_device())
        return gpu_name
    else:
        return "No GPU available."

print("GPU Name:", get_gpu_name())
```
